```python
#Hypothesis Testing

#Normally distributed samples

#H0:there is no relationship between wages and crime rates
#Ha:there is a (negative) between wages and crime rates


#Welch's T Test - two normally distributed samples - unequal variances

#H0: the means are equal
#Ha:the two groups do not have equal means

#Test

stats.ttest_ind(crime.Wage, crime.CrimeRate, equal_var=False).pvalue

#P-value = probability of getting your results by chance. We want this to be < 0.05

#If p-value is > 0.05 fail to reject H0
#If p-value is < 0.05 reject H0

```

```python
#Hypothesis Testing

#Non-normally distributed samples

#H0: there is no relationship between police expenditure and being in a Southern region
#Ha: there is a relationship between police expenditure and being in a Southern region

#Mann-Whitney U Test

#H0:both groups have the same median - the probability that a randomly drawn observation 
#from Expenditure is larger than a randomly drawn observation from Southern is equal to 50% (0.5)

#Ha: both groups do not have the same median

crime_south = crime[crime.Southern == 1]
crime_north = crime[crime.Southern == 0]

stat, p = stats.mannwhitneyu(crime_south.ExpenditureYear0, crime_north.ExpenditureYear0)

#Interpret

print('Statistics=%.3f, p=%.3f' % (stat,p))

alpha = 0.05

if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')


#P-value = probability of getting your results by chance. We want this to be < 0.05

#If p-value is > 0.05 fail to reject H0
#If p-value is < 0.05 reject H0
```

```python
#One sided
stats.mannwhitneyu(crime_south.ExpenditureYear0, crime_north.ExpenditureYear0, alternative='less').pvalue
```

```python
#Two-sided
stats.mannwhitneyu(crime_south.ExpenditureYear0, crime_north.ExpenditureYear0, alternative='two-sided').pvalue
```
