```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')

titanic
```

```python
titanic.head()
```

```python
#Pandas library
titanic.age.plot(kind='hist', bins=8)
```

```python
titanic.describe()
```

```python
titanic.info()
```

```python
#PLT library
plt.hist(titanic.age, bins=8)
```

```python
fig, ax = plt.subplots(ncols=2)
ax[0].hist(titanic.age, bins=8)
ax[0].set_ylabel('count')
ax[0].set_xlabel('age')
ax[1].hist(titanic.survived)
ax[1].set_title('Age Distribution on Titanic')
```

```python
import seaborn as sns
```

```python
sns.histplot(x=titanic.age, bins=8)
```

```python
sns.histplot(x=titanic.age, bins=8, kde=True, hue=titanic.survived, multiple='stack', palette='gnuplot2_r')
```

```python
sns.histplot(x=titanic.age, bins=8, kde=True, hue=titanic.sex, multiple='stack', palette='gnuplot2_r')
```

```python
sns.pairplot(titanic)
```

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(), annot=True)
```

```python
sns.histplot(x=titanic.pclass, bins=3, hue=titanic.survived, multiple='fill', discrete=True)
```

```python
sns.countplot(x=titanic.pclass, hue=titanic.survived, dodge=False)
```

```python
sns.displot(titanic, x=titanic['age'], col=titanic['sex'], row='embarked', binwidth=4, height=5, facet_kws=dict(margin_titles=True), kde=True) 
```
